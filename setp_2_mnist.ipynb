```python
from tensorflow.examples.tutorials.mnist import input_data
import tensorflow as tf
```

```python
mnist = input_data.read_data_sets("MNIST_data/", one_hot=True)
```

```python
x = tf.placeholder("float", shape=[None, 784])
y_ = tf.placeholder("float", shape=[None, 10])
```

```python
W = tf.Variable(tf.zeros([784,10]))
b = tf.Variable(tf.zeros([10]))
y = tf.nn.softmax(tf.matmul(x,W) + b)
```

```python
cross_entropy = -tf.reduce_sum(y_*tf.log(y))
train_op = tf.train.GradientDescentOptimizer(0.01).minimize(cross_entropy)
correct_prediction = tf.equal(tf.argmax(y,1), tf.argmax(y_,1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, "float"))
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for i in range(1000):
        batch = mnist.train.next_batch(50)
        sess.run(train_op, feed_dict={x: batch[0], y_: batch[1]})
        accuracy_ = sess.run(accuracy, feed_dict={x: mnist.test.images, y_: mnist.test.labels})
        print("%d epoch" % i, accuracy_)
```
