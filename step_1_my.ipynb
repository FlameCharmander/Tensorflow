```python
import tensorflow as tf
import numpy as np
```

```python
x_data = np.random.rand(2, 100) # (2 X 100)
np.shape(x_data)
```

```python
y_data = np.dot([[0.01, 0.02]], x_data) + 30.1 #(1 X 100)
np.shape(y_data)
```

```python
np.random.uniform(-1, 1, [1, 2])
```

```python
x = tf.placeholder(tf.float32, shape=[2, 100], name="input_x") #(2 X 100)
y = tf.placeholder(tf.float32, shape=[1, 100], name="input_y") #(1 X 100)
b = tf.Variable(tf.zeros([1]), name="variable_b") #(1,1)
W = tf.Variable(tf.random_uniform([1, 2], -1, 1), name="variable_W") #(2 X 1)
y_prediction = tf.add(tf.matmul(W, x, name="mul_y"),b, name="add_y") #(1 X 100)
```

```python
loss = tf.reduce_mean(tf.square(y - y_prediction)) #损失函数
optimizer = tf.train.GradientDescentOptimizer(0.5) #优化器
train_op = optimizer.minimize(loss) #训练
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for step in range(10000):
        loss_, _, W_, b_, _ = sess.run([loss, train_op, W, b, y_prediction], feed_dict={x:x_data, y:y_data})
        if step % 50 == 0:
            print("%d epoch" % step, loss_, W_, b_)
    writer = tf.summary.FileWriter("./my_graph", sess.graph)
    #writer.add_graph(sess.graph)
    writer.close()
```
